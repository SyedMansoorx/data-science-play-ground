```python
from sklearn.model_selection import train_test_split
```

```python
Z_train,Z_test,Y_train,Y_test= train_test_split(Z,Y, test_size=0.3, random_state=40)
```

```python
Y_predict_test=Linear_model.predict(Z_test)    #z -- input
```
